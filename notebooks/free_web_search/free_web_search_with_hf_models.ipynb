# Free Web Search with Hugging Face Models (v13.0.0)

This notebook demonstrates how to use [`free-web-search-ultimate`](https://github.com/wd041216-bit/free-web-search-ultimate) — a **zero-cost, API-key-free** web search MCP server — together with Hugging Face models.

Unlike paid search APIs (Tavily, Brave, Exa), `free-web-search-ultimate` requires no API key and supports:
- **Multiple engines**: DuckDuckGo, Bing, Google, Brave, Yahoo
- **Content types**: web, news, images, videos
- **MCP protocol**: compatible with any MCP-enabled LLM client

```python
from free_web_search_ultimate import UltimateSearcher

# Initialize the searcher — no API key needed!
searcher = UltimateSearcher()

# Search the web
results = searcher.search('Hugging Face latest models 2025', engine='duckduckgo', max_results=5)
for r in results:
    print(f"Title: {r['title']}")
    print(f"URL: {r['url']}")
    print(f"Snippet: {r['snippet'][:100]}...")
    print()
```

## Using with Hugging Face Inference API

Combine free web search with any Hugging Face model for grounded generation.

```python
# Search for news
news = searcher.search_news('AI research breakthroughs', max_results=3)
for article in news:
    print(f"[{article.get('date', 'N/A')}] {article['title']}")
    print(f"  Source: {article['url']}")
    print()
```
